```python
from svpg.utils.utils import load_algo
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")

base_dir = "../runs/MyCartPoleSwingUp-v0/22-05-11/"
```

```python
coefs = [1e-3, 1e-2, 0.1, 0.5, 1]
dirs = ["12-46-54", "13-55-54", "14-15-25", "14-32-22", "14-50-41"]
```

```python
plt.figure(figsize=(9,6))

for i, d in enumerate(dirs):
    _, _, rewards, steps = load_algo(base_dir + d + "/A2C-independent")
    rewards = rewards.mean(axis=0)
    plt.plot(steps, rewards, linewidth=2, label=f"entropy_coef = {coefs[i]}")

plt.legend(loc="lower right")
plt.plot()
```
