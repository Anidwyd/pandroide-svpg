```python
from svpg.utils.utils import load_algo_data
from svpg.visu.state_visitation import plot_state_visitation

from omegaconf import OmegaConf
```

```python
config = OmegaConf.create({
    "algorithm": {
        "seed": 4,
        "n_evals": 100,
    },
    "gym_env": {
        "classname": "svpg.agents.env.make_gym_env",
        "env_name": "CartPole-v1",
        "max_episode_steps": 500,
    }
})
```

```python
env_name = "CartPole-v1"
directory = f"../archives/{env_name}/22-05-07/11-39-36/"

a2c_agents, _, a2c_rewards = load_algo_data(directory + "A2C-independant")
svpg_agents, _, svpg_rewards = load_algo_data(directory + "A2C-SVPG")
svpga_agents, _, svpga_rewards = load_algo_data(directory + "A2C-SVPG_annealed")
```

```python
plot_state_visitation(config, a2c_agents, a2c_rewards, "A2C", cmap="Blues", save=False)
```

```python
plot_state_visitation(config, svpg_agents, svpg_rewards, "SVPG", cmap="Reds", save=False)
```

```python
plot_state_visitation(config, svpga_agents, svpga_rewards, "SVPG-ANNEALED", cmap="Greens", save=False)
```
